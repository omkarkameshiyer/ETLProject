```python
import pandas as pd
from sqlalchemy import create_engine
import json

credit_csv = "tmdb_5000_credits.csv"
credit_df = pd.read_csv(credit_csv)
credit_df.head()


movie_csv_file = "tmdb_5000_movies.csv"
movie_data_df = pd.read_csv(movie_csv_file)
movie_data_df.head()

title = pd.DataFrame(credit_df["title"])
title.set_index("title").head()
```

```python
castList = []
for index, row in credit_df.iterrows():
   cast_list = json.loads(row["cast"])
   cast_per_movie = []
   for cast in cast_list:
       cast_per_movie.append(cast['name'])
   castList.append(cast_per_movie)
```

```python
type(castList)
df_cast = pd.DataFrame({'col':castList})
```

```python
df_cast.head()
len(df_cast)
```

```python
credit_df['castList'] = df_cast
```

```python
credit_df.head()
```

```python
crews = pd.DataFrame(credit_df["crew"])
```

```python
castList = []
for index, row in credit_df.iterrows():
   cast_list = json.loads(row["crew"])
   cast_per_movie = []
   for cast in cast_list:
       if(cast["job"] == "Director") :
           cast_per_movie.append(cast["name"])
   castList.append(cast_per_movie)
```

```python
len(castList)
df_dir = pd.DataFrame({'col':castList})
```

```python
credit_df['director'] = df_dir
```

```python
credit_df.head()
```

```python
credit_df = credit_df.drop("cast", axis=1)
credit_df = credit_df.drop("crew", axis=1)
```

```python
credit_df.head()
```

```python

movie_data_df = movie_data_df.drop("original_language", axis=1)
movie_data_df = movie_data_df.drop("budget", axis=1)
movie_data_df = movie_data_df.drop("production_companies", axis=1)
movie_data_df = movie_data_df.drop("production_countries", axis=1)
movie_data_df = movie_data_df.drop("runtime", axis=1)
```

```python
movie_data_df = movie_data_df.drop("genres", axis=1)
movie_data_df = movie_data_df.drop("popularity", axis=1)
movie_data_df = movie_data_df.drop("spoken_languages", axis=1)


movie_data_df.head()
```

```python
movie_data_df = movie_data_df.drop("status", axis=1)
movie_data_df = movie_data_df.drop("vote_average", axis=1)
movie_data_df = movie_data_df.drop("vote_count", axis=1)


movie_data_df.head()
```
